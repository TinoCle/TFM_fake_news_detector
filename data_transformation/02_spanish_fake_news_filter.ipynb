# Testing with a loops over the df

Full text of the news are in the column 'content'

Test if there are news with 2 languages.

Finally detect which news are in spanish and load them in another DataFrame

### First iteration

1) We know there are news in spanish, so lets test how many are in a sample of 30000.

2) If there are news in spanish, print the row index.

3) Checkout the rows with news in spanish and lets see if there are 2 languages.

```python
%%time

import pandas as pd
from langdetect import detect
from langdetect import detect_langs

count = 0

df = pd.read_csv('news_csv_filtered.csv', sep = ',', nrows = 20000)

for index, row in df.iterrows():
    if detect(row['content']) == 'es':
        count += 1
        langs = len(detect_langs(row['content'])) #Count the number of languages detected in the text
        print('The row %d has %d languages' % (index, langs))
    else:
        pass

print('There are %d news in spanish' % count)
```

```python
#Checking out the row 10895 with 2 languages

print(detect_langs(df['content'].iloc[10895]))

print(df['content'].iloc[10895])
```

### #Code1: With iterrows and dropping the rows that don't have news in spanish language and have more than one language

```python
%%time

df_esp1 = pd.read_csv('news_csv_filtered.csv', sep = ',', nrows = 20000)

for index, row in df_esp1.iterrows():
    # Detect if the news are in spanish and only those with one language detected
    if not(detect(row['content']) == 'es' and len(detect_langs(row['content'])) == 1):
        df_esp1.drop(index, inplace = True)
    else:
        pass
```

### #2 Code:  With iterrows, appending rows with news in spanish in a list, and finally create a new DataFrame

```python
%%time

row_esp = []

df = pd.read_csv('news_csv_filtered.csv', sep = ',', nrows = 20000)

for index, row in df.iterrows():
    # Detect if the news are in spanish and only those with one language detected
    if detect(row['content']) == 'es' and len(detect_langs(row['content'])) == 1:
        row_esp.append(row.values)
    else:
        pass

df_esp2 = pd.DataFrame(row_esp, columns = df.columns)
```

### #3 Code:  Without iterrows, appending each value in a list, create a new DataFrame with all the lists created.

```python
%%time

id_list = []
type_list = []
title_list = []
content_list = []
url_list = []
domain_list = []

num = 20000

df = pd.read_csv('news_csv_filtered.csv', sep = ',', nrows = num)

for n in range(num):
    if detect(df['content'].iloc[n]) == 'es' and len(detect_langs(df['content'].iloc[n])) == 1:
        content_list.append(df['content'].iloc[n])
        id_list.append(df['id'].iloc[n])
        type_list.append(df['type'].iloc[n])
        title_list.append(df['title'].iloc[n])
        url_list.append(df['url'].iloc[n])
        domain_list.append(df['domain'].iloc[n])
    else:
        pass

df_esp3 = pd.DataFrame({
    'id': id_list,
    'type': type_list,
    'domain': domain_list,
    'title': title_list ,
    'content': content_list,
    'url': url_list
    })
```

### #1 Code with chunks

```python
%%time

df = pd.read_csv('news_csv_filtered.csv', sep = ',', lineterminator = '\n', nrows = 20000, chunksize = 1000)

all_chunks1 = pd.DataFrame()

for i, chunk in enumerate(df):
    print(i, len(chunk))

    for index, row in chunk.iterrows():
        if not(detect(row['content']) == 'es' and len(detect_langs(row['content'])) == 1):
            chunk.drop(index, inplace = True)
        else:
            pass
    
    all_chunks1 = all_chunks1.append(chunk)
```

### #3 Code with chunks

```python
%%time

id_list = []
type_list = []
domain_list = []
title_list = []
content_list = []
url_list = []

df = pd.read_csv('news_csv_filtered.csv', sep = ',', lineterminator = '\n', nrows = 20000, chunksize = 1000)

for i, chunk in enumerate(df):
    print(i, len(chunk))
    num = len(chunk.index)

    for n in range(num):
        if detect(chunk['content'].iloc[n]) == 'es' and len(detect_langs(chunk['content'].iloc[n])) == 1:
            id_list.append(chunk['id'].iloc[n])
            type_list.append(chunk['type'].iloc[n])
            domain_list.append(chunk['domain'].iloc[n])
            title_list.append(chunk['title'].iloc[n])            
            content_list.append(chunk['content'].iloc[n])
            url_list.append(chunk['url'].iloc[n])
        else:
            pass

all_chunks3 = pd.DataFrame({
    'id': id_list,
    'type': type_list,
    'domain': domain_list,
    'title': title_list ,
    'content' : content_list,
    'url' : url_list,
    })
```

Langdetect package can experiments difficulties if a row lacks text or only includes symbols/numbers, even if they are string type.

```python
%%time

import pandas as pd
from langdetect import detect
from langdetect import detect_langs
from datetime import datetime

row_spanish = []

df = pd.read_csv('news_csv_filtered.csv', sep = ',', lineterminator = '\n', chunksize = 100000)

for i, chunk in enumerate(df):
    print("Start with chunk %d : %d at %s" % (i, len(chunk), datetime.now()))

    for index, row in chunk.iterrows():
        if detect(row['content']) == 'es' and len(detect_langs(row['content'])) == 1:
            row_spanish.append(row.values)
        else:
            pass

df_spanish = pd.DataFrame(row_spanish, columns = chunk.columns)
```
